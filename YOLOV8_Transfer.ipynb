```python
model = YOLO("yolov8n.yaml")
model = YOLO("yolov8n.pt")
print(model.model)
```

```python
model.predict("test-images/shiba_inu_6.jpg",save=True)

# URL of the image
image_url = r'runs/detect/predict2/shiba_inu_6.jpg'

img = cv2.imread(image_url)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
